```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# ==============================================================
# 2. CARGA Y PREPROCESAMIENTO DE DATOS
# ==============================================================
df = pd.read_csv("USAHousingDataset.csv")

# Limpieza y transformación básica
df = df.drop(columns=['date', 'street', 'statezip', 'country'])
df = df[df['price'] > 0]
df = pd.get_dummies(df, columns=['city'])

# Variables predictoras y objetivo
X = df.drop(columns=['price'])
y = np.log(df['price'])  # Escala logarítmica para estabilizar la varianza

# División de los datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
train_df = X_train.copy()
train_df["price"] = y_train

test_df = X_test.copy()
test_df["price"] = y_test
```

```python
train_df.to_csv("housing_train.csv", index=False)
test_df.to_csv("housing_test.csv", index=False)
```
